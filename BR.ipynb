##Carrega dados

```
import pandas as pd

data = pd.read_excel('/content/drive/MyDrive/dados.xlsx')

df_br = pd.DataFrame(data)
```

## Seleciona colunas pelo ano 2017

```
df_br.reset_index()

columns = ['Data de Fim da Consulta', 'Data de Início da Consulta', 'Link do Jogo','Turno','Estádio','Cidade' ]
df_br = df_br.drop(columns, axis=1)

df_br17_mandante = df_br[((df_br['Temporada'] == 2017) & (df_br['Série'] == 'A'))][['Temporada', 'Mandante','Visitante','Gol Mandante','Gol Visitante','Série','Rodada','Resultado']]
df_br17_visitante = df_br[((df_br['Temporada'] == 2017) & (df_br['Série'] == 'A'))][['Temporada','Mandante','Visitante','Gol Mandante','Gol Visitante','Série','Rodada','Resultado']]

```

## Calcula Resultado das partidas 

```
def result(x, y):
    if x < y: 
      return 0
    elif x > y:
      return 3
    else:
      return 1

df_br17_mandante['Resultado'] = df_br17_mandante.apply(lambda row: result(row['Gol Mandante'], row['Gol Visitante']), axis=1)
df_br17_visitante['Resultado'] = df_br17_visitante.apply(lambda row: result(row['Gol Visitante'], row['Gol Mandante']), axis=1)
frames =[df_br17_mandante ,df_br17_visitante ]
result = pd.concat(frames, keys=["Mandante", "Visitante"])
```

## Reseta o index

```
result = result.sort_index()
result = result.reset_index()
```
